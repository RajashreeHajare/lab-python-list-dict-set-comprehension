# Lab | List, Dict and Set Comprehension

````python

def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

    
def get_customer_orders(products):
    # Prompt the user to enter the number of customer orders
    num_orders = int(input("Enter the number of customer orders: "))
    
    # Gather the product names using comprehension
    customer_orders = {input(f"Enter the name of a product that a customer wants to order ({i + 1}/{num_orders}): ").strip().lower()
                       for i in range(num_orders)}
    
    # Validate and keep only products in the list
    customer_orders = {order for order in customer_orders if order in products}
    
    return customer_orders

def update_inventory(customer_orders, inventory):
    # Update the inventory dictionary using comprehension
    inventory = {product: (quantity - 1) for product, quantity in inventory.items()
                 if product in customer_orders and quantity > 0}
    # Remove products with zero quantity
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    return inventory

    
def manage_customer_orders():
    # Define the list of products
    products = ["t-shirt", "mug", "hat", "book", "keychain"]

    # Initialize inventory
    inventory = initialize_inventory(products)

    # Get customer orders
    customer_orders = get_customer_orders(products)

    # Calculate order statistics
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100

    # Print order statistics
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Unique Products Ordered: {percentage_ordered:.2f}%")

    # Update inventory based on customer orders
    inventory = update_inventory(customer_orders, inventory)

    # Print updated inventory
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")


manage_customer_orders()
````
